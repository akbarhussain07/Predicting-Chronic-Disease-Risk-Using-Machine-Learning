```python
with open('user_behavior_dataset_to_write.csv', mode='w') as write_file:
    my_writer = csv.writer(write_file, delimiter=',', quotechar='"')
    
    for i in range(0,len(u_list)):
        my_writer.writerow(u_list[i])
```

```python
#%% read and write
with open('user_behavior_dataset_to_write.csv', mode='a+') as write_file:
    employee_writer = csv.writer(write_file, delimiter=',', quotechar='"')
    
    with open('user_behavior_dataset.csv', mode ='r') as file:
        csvFile = csv.reader(file)
        for lines in csvFile:
            employee_writer.writerow(lines)

#%% Using pandas to read as write

user_data = pd.read_csv('user_behavior_dataset.csv')

#%%
user_data.to_csv('out.csv', index=False)
```
