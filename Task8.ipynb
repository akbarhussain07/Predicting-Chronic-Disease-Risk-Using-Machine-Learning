```python
import numpy as np 
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression

df = pd.read_csv(r"D:\MYFolder( Extra Effort)\DATASETS\user_behavior_dataset.csv")
```

```python
df.head()
```

```python
#Predict Data Usage

columns = ['App Usage Time (min/day)', 'Battery Drain (mAh/day)',
       'Number of Apps Installed', 'Data Usage (MB/day)', 'Age']

data = df[columns]
```

```python
data.head()
```

```python
#%%Separate Features and test-train data

feature_name = ['App Usage Time (min/day)', 'Battery Drain (mAh/day)',
       'Number of Apps Installed', 'Age']
target = ['Data Usage (MB/day)']
```

```python
X = data[feature_name]
y = data[target]
```

```python
X
```

```python
y
```

```python


test_data = 0.2
seed = 10
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_data, random_state=seed)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_test.shape    
```

```python
X_test.shape
```

```python
#%% Train the Regression Model
regr = LinearRegression() 
regr.fit(X_train, y_train)
```

```python
#%% Find predictions from trained model
y_pred = regr.predict(X_test) 
```

```python
y_test
```

```python
y_pred
```

```python
result=y_pred-y_test
result
```

```python
std_dev=np.std(result)
```

```python
#%% Compute the Errors
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error
  
mae = mean_absolute_error(y_true=y_test, y_pred=y_pred) 
#squared True returns MSE value, False returns RMSE value. 
mse = mean_squared_error(y_true=y_test, y_pred=y_pred) #default=True 
rmse = root_mean_squared_error(y_true=y_test, y_pred=y_pred) 
  
print("MAE:",mae) 
print("MSE:",mse) 
print("RMSE:",rmse)
# print("StandardDeviation:",std_dev)
```

```python

#%% K Fold Cross Validation

from sklearn.model_selection import cross_val_score
regr = LinearRegression()
```

```python
n_splits = 10
scores = cross_val_score(regr, X, y, scoring='neg_root_mean_squared_error', cv=n_splits)
print("Score: ",scores)
rmse_scores = -scores
print("Scores:", rmse_scores)
print("Mean:", rmse_scores.mean())
print("StandardDeviation:", rmse_scores.std())
```

```python
#%% Logistic Regression (Classification)
#Predict User Behavior Class
columns = ['App Usage Time (min/day)', 'Battery Drain (mAh/day)',
       'Number of Apps Installed', 'Data Usage (MB/day)', 'Age', 'User Behavior Class']

data = df[columns]
```

```python
feature_name = ['App Usage Time (min/day)', 'Battery Drain (mAh/day)',
       'Number of Apps Installed', 'Age', 'Data Usage (MB/day)']
target = ['User Behavior Class']
```

```python
data['User Behavior Class'].value_counts()
```

```python
X = data[feature_name]
y = data[target]

test_data = 0.2
seed = 10
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_data, random_state=seed)
```

```python
#%% Train the Regression Model
classifier = LogisticRegression() 
classifier.fit(X_train, y_train)
```

```python
#%% Find predictions from trained model
y_pred = classifier.predict(X_test) 
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print("Logistic Regression model accuracy(in %):",
      accuracy_score(y_test, y_pred)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
