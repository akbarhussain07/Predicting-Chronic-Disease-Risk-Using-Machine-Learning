```python
# Linear Regression

import numpy as np 
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
datas = pd.read_csv('data.csv')
```

```python
X = datas.iloc[:, 1:2].values
y = datas.iloc[:, 2].values
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
 
lin.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
 
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

#%% Polynomial regression
#poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)
```

```python
import matplotlib.pyplot as plt
 
# Visualising the Linear Regression results
plt.scatter(X, y, color='blue')

y_predict = lin.predict(X)

plt.plot(X, y_predict, color='red')
plt.title('Linear Regression')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
 
plt.show()
```

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y, color='blue')
 

y_predict = lin2.predict(X_poly)
plt.plot(X, y_predict,
         color='red')
plt.title('Polynomial Regression')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
 
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\MYFolder( Extra Effort)\DATASETS\user_behavior_dataset.csv")
```

```python
#%%
#Predict Data Usage

columns = ['App Usage Time (min/day)', 'Battery Drain (mAh/day)',
       'Number of Apps Installed', 'Data Usage (MB/day)', 'Age']

data = df[columns]
```

```python
#%%Separate Features and test-train data

feature_name = ['App Usage Time (min/day)', 'Battery Drain (mAh/day)',
       'Number of Apps Installed', 'Age']

target = ['Data Usage (MB/day)']

X = data[feature_name]
y = data[target]

test_data = 0.2
seed = 11
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_data, random_state=seed)
```

```python
#%% Train the Linear Regression Model
regr = LinearRegression() 
regr.fit(X_train, y_train) 
```

```python
#%% Find predictions from trained model
y_pred = regr.predict(X_test)
```

```python
#%% Compute the Errors
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error
  
mae = mean_absolute_error(y_true=y_test, y_pred=y_pred) 
#squared True returns MSE value, False returns RMSE value. 
mse = mean_squared_error(y_true=y_test, y_pred=y_pred) #default=True 
rmse = root_mean_squared_error(y_true=y_test, y_pred=y_pred) 
  
print("MAE:",mae) 
print("MSE:",mse) 
print("RMSE:",rmse)
```

```python
#%% K Fold Cross Validation
from sklearn.model_selection import cross_val_score
regr = LinearRegression()
```

```python


n_splits = 5
scores = cross_val_score(regr, X, y, scoring='neg_root_mean_squared_error', cv=n_splits)

rmse_scores = -scores
print("Scores:", rmse_scores)
print("Mean:", rmse_scores.mean())
print("StandardDeviation:", rmse_scores.std())
```
